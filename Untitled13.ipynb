```python
import pandas as pd
import re

# CSV dosyasını oku
df = pd.read_csv("universite_ders_verisi_10000_corrected_final_gg.csv")

# Sadece metin sütunlarını seç
text_columns = df.select_dtypes(include='object').columns

# 1. Baştaki ve sondaki boşlukları sil
for col in text_columns:
    df[col] = df[col].str.strip()

# 2. Küçük harfe çevir
for col in text_columns:
    df[col] = df[col].str.lower()

# 3. Karakter düzeltmeleri
def fix_turkish_chars(text):
    if isinstance(text, str):
        text = text.replace("Ã", "ö").replace("Ã¶", "ö").replace("Ãœ", "ü").replace("Ã¼", "ü")
        text = text.replace("Ä", "ğ").replace("ÄŸ", "ğ").replace("Å", "ş").replace("Å", "ş")
        text = text.replace("Ã§", "ç").replace("Ã‡", "ç").replace("Ä±", "ı")
    return text

for col in text_columns:
    df[col] = df[col].apply(fix_turkish_chars)

# 4. Özel karakter temizliği
def clean_special_chars(text):
    if isinstance(text, str):
        text = re.sub(r"[^a-zA-Z0-9çÇşŞğĞüÜöÖıİ\s]", "", text)
    return text

for col in text_columns:
    df[col] = df[col].apply(clean_special_chars)

# 5. Boş hücreleri kontrol et
print("Boş hücre sayısı:\n", df.isnull().sum())

# 6. Yinelenen satırları sil
df.drop_duplicates(inplace=True)

# 7. Sonucu kaydet
df.to_csv("temizlenmis_veri.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Örnek veri: kelimeler ve sıklıkları
# Kelime sıklıkları bir dictionary formatında olmalı, örneğin:
# word_frequencies = {'kelime1': 1000, 'kelime2': 800, 'kelime3': 600, ...}

word_frequencies = {
    'kelime1': 1000,
    'kelime2': 800,
    'kelime3': 600,
    'kelime4': 400,
    'kelime5': 350,
    'kelime6': 300,
    'kelime7': 250,
    'kelime8': 200,
    'kelime9': 180,
    'kelime10': 150
}

# Sıklıkları sıralayın (azalan sırayla)
sorted_frequencies = sorted(word_frequencies.values(), reverse=True)

# Kelime sırasını (X ekseni) oluşturun
ranks = np.arange(1, len(sorted_frequencies) + 1)

# Y ekseni: Sıklıkların logaritması
log_frequencies = np.log(sorted_frequencies)

# X ekseni: Kelime sıralarının logaritması
log_ranks = np.log(ranks)

# Grafik çizimi
plt.figure(figsize=(8, 6))
plt.plot(log_ranks, log_frequencies, marker='o', linestyle='-', color='b')

# Grafik etiketleri
plt.title("Zipf Yasasına Göre Log-Log Grafiği", fontsize=14)
plt.xlabel("Log(Kelime Sırası)", fontsize=12)
plt.ylabel("Log(Sıklık)", fontsize=12)

# Grafik gösterimi
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from tqdm import tqdm
import zeyrek

# Gerekli indirmeler
nltk.download('punkt')
nltk.download('stopwords')

# CSV dosyasını oku (kendi dosya adınla değiştir)
df = pd.read_csv("universite_ders_verisi_10000_corrected_final_gg.csv")

# Zeyrek lemmatizer ve stopword'ler
lemmatizer = zeyrek.MorphAnalyzer()
stop_words = set(stopwords.words('turkish'))

# Lemmatization fonksiyonu
def lemmatize_text(text):
    tokens = word_tokenize(str(text))  # Metni kelimelere ayır
    tokens = [token.lower() for token in tokens if token.isalpha() and token.lower() not in stop_words]
    
    lemmas = []
    for token in tokens:
        analysis = lemmatizer.analyze(token)
        if analysis and analysis[0][0][1]:  # Lemma bilgisi varsa
            lemmas.append(analysis[0][0][1])
        else:
            lemmas.append(token)  # Yoksa kelimeyi olduğu gibi ekle
    
    return " ".join(lemmas)

# Lemma işlemini uygula (Ders Tanımı sütununa)
tqdm.pandas()
df["lemmatized_text"] = df["Ders Tanımı"].progress_apply(lemmatize_text)

# Yeni dosyaya kaydet
df.to_csv("lema_sonuclari.csv", index=False)
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# CSV dosyasını oku
df = pd.read_csv("lema_sonuclari.csv")

# "lemmatized_text" sütununu al ve tokenleştir
lemmatized_tokens = " ".join(df["lemmatized_text"]).split()

# Frekansları hesapla
lemma_counts = Counter(lemmatized_tokens)

# En sık geçen 10 kelimeyi al
common_lemmas = lemma_counts.most_common(10)

# Zipf grafiğini çiz
lemmas, counts = zip(*common_lemmas)

plt.figure(figsize=(10, 6))
plt.bar(lemmas, counts)
plt.title("En Sık Geçen 10 Lemma")
plt.xlabel("Lemma")
plt.ylabel("Frekans")
plt.xticks(rotation=45)

# Grafiği dosyaya kaydet
plt.savefig("zipf_graph.png", bbox_inches='tight')

# Grafiği göster
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Lemmatized metinleri içeren dosyayı oku
df = pd.read_csv("lema_sonuclari.csv")  # Dosya yolunu doğru belirtin

# "lemmatized_text" sütununu al ve tek bir string haline getir
lemmatized_texts = " ".join(df["lemmatized_text"])

# Kelimeleri ayır ve frekanslarını hesapla
words = lemmatized_texts.split()
word_counts = Counter(words)

# Kelimeleri frekansa göre sıralayalım
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Sıklıkların (frekansların) ve sıralamanın listesi
frequencies = [count for word, count in sorted_word_counts]
ranks = np.arange(1, len(frequencies) + 1)

# Zipf yasası grafiği oluştur
plt.figure(figsize=(10, 6))
plt.plot(ranks, frequencies, marker="o", linestyle="-", color="b")
plt.xscale('log')
plt.yscale('log')
plt.title("Zipf's Law - Lemmatized Data")
plt.xlabel("Rank (Log scale)")
plt.ylabel("Frequency (Log scale)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Grafiği kaydet
plt.savefig("zipf_law_lemmatized.png")

# Grafiği göster
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# NLTK için gerekli olanları indir
nltk.download('punkt')
nltk.download('stopwords')

# Stopwords yükle
stop_words = set(nltk.corpus.stopwords.words('turkish'))

# Stemmer'ı başlat
stemmer = PorterStemmer()

# CSV dosyasını oku
df = pd.read_csv("universite_ders_verisi_10000_corrected_final_gg.csv")

# Metni al (Ders Tanımı sütunu)
texts = df['Ders Tanımı'].dropna().tolist()

# Stemleme fonksiyonu
def stem_text(text):
    tokens = word_tokenize(text)  # Kelimelere ayır
    filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    return " ".join(stemmed_tokens)

# Tüm metni stemleyelim
stemmed_texts = [stem_text(text) for text in texts]

# Stemlenmiş metni yeni bir CSV dosyasına kaydedelim
df['stemmed_text'] = stemmed_texts
df.to_csv("stemmed_texts.csv", index=False)

print("Stemleme işlemi tamamlandı ve kaydedildi.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Stemlenmiş veriyi oku
df = pd.read_csv("stemmed_texts.csv")

# "stemmed_text" sütununu al
stemmed_texts = df['stemmed_text'].dropna().tolist()

# Stemlenmiş metni birleştirip kelimeleri ayıralım
stemmed_text_combined = " ".join(stemmed_texts)
words = stemmed_text_combined.split()

# Kelime frekanslarını hesaplayalım
word_counts = Counter(words)

# Kelimeleri sıklıklarına göre sırala
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Frekansları ve sıralamaları al
frequencies = [count for word, count in sorted_word_counts]
ranks = np.arange(1, len(frequencies) + 1)

# Zipf yasası grafiği
plt.figure(figsize=(10, 6))
plt.plot(ranks, frequencies, marker="o", linestyle="-", color="b")
plt.xscale('log')
plt.yscale('log')
plt.title("Zipf's Law - Stemmed Data")
plt.xlabel("Rank (Log scale)")
plt.ylabel("Frequency (Log scale)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Grafiği kaydet
plt.savefig("zipf_law_stemmed.png")

# Grafiği göster
plt.show()

print("Zipf yasası grafiği kaydedildi.")
```

```python
import pandas as pd

# Dosyayı oku
df_lemmatized = pd.read_csv(r"C:\Users\Casper\lema_sonuclari.csv")

# Boş olmayan verileri al ve ilk birkaç satırı kontrol et
df_lemmatized = df_lemmatized.dropna(subset=['lemmatized_text'])
print(df_lemmatized.head())

# Cümleleri kelime listelerine çevir (tokenization)
lemmatized_sentences = df_lemmatized['lemmatized_text'].apply(lambda x: x.split()).tolist()

# Örnek çıktı
print(lemmatized_sentences[:3])  # İlk 3 cümleyi göster
```

```python
from gensim.models import Word2Vec
import os

# Parametre seti
parameters = [
    {'model_type': 'cbow', 'window': 2, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 2, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 300},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 300}
]

# Modelleri kaydedeceğimiz klasör
output_dir = r"C:\Users\Casper\word2vec_models"
os.makedirs(output_dir, exist_ok=True)

# Eğitim döngüsü
for param in parameters:
    sg = 0 if param['model_type'] == 'cbow' else 1
    model = Word2Vec(
        sentences=lemmatized_sentences,
        vector_size=param['vector_size'],
        window=param['window'],
        sg=sg,
        workers=4,
        min_count=1,
        epochs=20
    )

    # Dosya adı oluşturma
    model_name = f"lemmatized_model_{param['model_type']}_window{param['window']}_dim{param['vector_size']}.model"
    model_path = os.path.join(output_dir, model_name)

    # Modeli kaydet
    model.save(model_path)
    print(f"Model kaydedildi: {model_name}")
```
